# Load Libraries

```python
import pandas as pd
import numpy as np

!pip install pycodestyle pycodestyle_magic flake8
%load_ext pycodestyle_magic

from IPython.display import display, HTML
```

# Load data

```python
# SNP data
snpfile = '~/Data/SNP/CCLE.txt' 
snp = pd.read_csv(snpfile, sep='\t')
print(snp)

# eGEM Entrez IDs
entrezfile = '~/Data/IDs/eGEM_entrez.txt'
eGEM_entrez = pd.read_csv(entrezfile)
eGEM_entrez = eGEM_entrez.drop_duplicates(keep='first')
print(eGEM_entrez)
```

# Filter SNPs

```python
filtered_snps = pd.merge(eGEM_entrez, snp, 
    how='inner', left_on='Entrez', right_on='Entrez_Gene_Id'
)
filtered_snps = filtered_snps.drop('Entrez', axis=1)
filtered_snps = filtered_snps.drop_duplicates(keep='first')

pd.options.display.max_columns = None
display(filtered_snps)
```

```python
subset = filtered_snps[['Hugo_Symbol', 'Entrez_Gene_Id', 
                        'Variant_Classification', 'Variant_Type',
                        'isDeleterious', 'isTCGAhotspot', 'isCOSMIChotspot', 
                        'Broad_ID', 'Tumor_Sample_Barcode']]
display(subset)
variants = np.unique(subset['Variant_Classification'])
```

# Label encode 

```python
for var in variants:
    subset[var] = subset['Variant_Classification'] == var
display(subset)
```

```python
cols_to_subset = np.append(variants, ["isDeleterious", "Hugo_Symbol",
                                  "isTCGAhotspot", "isCOSMIChotspot",  
                                  "Broad_ID", "Tumor_Sample_Barcode"])
variant_subset = subset[cols_to_subset]
display(variant_subset)
```

# Aggregate based on cell line and genes

```python
counts_df = variant_subset.groupby(by=['Hugo_Symbol', 
                                       'Broad_ID', 
                                       'Tumor_Sample_Barcode']).sum()
display(counts_df)
print(counts_df.max())
```

```python
cols_to_subset = np.append(variants, ["isDeleterious", "Hugo_Symbol",
                                  "isTCGAhotspot", "isCOSMIChotspot"])
variant_subset = subset[cols_to_subset]
counts_df = variant_subset.groupby(by=['Hugo_Symbol']).sum()
display(counts_df)
print(counts_df.max())
colnames = list(np.unique(counts_df.columns))
```

```python
#!pip install seaborn
#!pip install plotly
import seaborn as sns
import plotly
import plotly.graph_objects as go

#print(counts_df.shape)
tmp = counts_df.sort_values(by=['Missense_Mutation'], ascending=True)
display(tmp)
# Static
#sns.heatmap(tmp)

# Dynamic
fig = go.Figure(
    data=go.Heatmap(
        z=tmp,
        x=colnames,
        y=list(tmp.index)
    )
)

fig.update_layout(
    autosize=False,
    width=1600,
    height=800
)
    
fig.show()
```
